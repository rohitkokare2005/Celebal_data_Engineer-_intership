## 3. Handle Missing Values

```python
data.isnull().sum()
```

```python

data['discount'] = data['discount'].fillna(0)

# Verify that missing values in 'discount' are handled
print(f"Remaining missing values in discount: {data['discount'].isnull().sum()}")
```

## 4 . Perform Basic Operations: Filtering & Selection 

```python


# Filter: Top-rated products (rating >= 4.5)
top_rated = data[data['rating'] >= 4.5]

print(f"Products with rating >= 4.5: {len(top_rated)}")

# Display the first 10 results
display(top_rated[['title', 'category', 'rating', 'final_price']].head(10))
```

```python
# Filter: High discount products (discount >= 50%)
high_discount = data[data['discount'] >= 50]
print(f'Products with discount >= 50%: {len(high_discount)}')
display(high_discount[['title', 'initial_price', 'final_price', 'discount']].head(10))
```

## 5 . Remove Duplicates

```python
initial_rows = data.shape[0]

# Remove duplicate entries based on unique product IDs
data = data.drop_duplicates(subset=['product_id'])

final_rows = data.shape[0]
print(f"✅ Deduplication complete! Removed {initial_rows - final_rows} duplicate records.")
```

## 6 . Feature Engineering & Derived Columns

```python

data['final_price'] = data['final_price'].astype(str).str.replace(r'[₹,\"\s]', '', regex=True)
data['final_price'] = pd.to_numeric(data['final_price'], errors='coerce')

data['quantity'] = 1

data['total_amount'] = data['final_price'] * data['quantity']

print("✅ Feature engineering completed successfully! Sample columns:")
display(data[['title', 'final_price', 'quantity', 'total_amount']].head())
```

## 7 . Save Cleaned Dataset

```python
output_file = "cleaned_product_dataset.csv"
data.to_csv(output_file, index=False)
print(f"🎉 Cleaned dataset saved successfully as: '{output_file}'")
```
